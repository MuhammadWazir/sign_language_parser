```python
from ultralytics import YOLO

# Load the trained model
model = YOLO('runs/detect/train/weights/best.pt')

# Perform inference on an image (or video)
# Replace 'path_to_image_or_video' with the path to the image or video file
results = model('/content/drive/MyDrive/datasets/test/images/A22_jpg.rf.2584deaa261bba945ec5beb1436bed69.jpg')


# To display the results
for result in results:
    result.show()

    boxes = result.boxes
    class_ids = boxes.cls  # Class IDs for the detected objects

    # Get the class names for each detected object
    class_names = [result.names[int(class_id)] for class_id in class_ids]

    # Print the detected class names
    print("Detected object classes:", class_names)
```
